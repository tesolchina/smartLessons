```python
# Write code to create bar charts for perceived fairness, government_consideration, policy_helpfulness, and waste_severity with correct Likert scales and labels

plt.figure(figsize=(16, 12))

# Define variables, titles, and Likert scale options for each
variables = ['fairness', 'government_consideration', 'policy_helpfulness', 'waste_severity']
titles = [
    'Perceived Fairness of Policy',
    'Perceived Government Consideration',
    'Perceived Policy Helpfulness',
    'Perceived Severity'
]
orders = [
    [1, 2, 3, 4, 5],  # fairness: 1=Very unfair, 5=Very fair
    [1, 2, 3, 4, 5],  # government_consideration: 1=Not considered at all, 5=Considered very fully
    [1, 2, 3, 4],     # policy_helpfulness: 1=Not helpful at all, 4=Very helpful (no neutral)
    [1, 2, 3, 4]      # waste_severity: 1=Not at all severe, 4=Highly severe (no neutral)
]
labels = [
    ['Very unfair', 'Unfair', 'Neutral', 'Fair', 'Very fair'],
    ['Not considered at all', 'Not fully considered', 'Somewhat considered', 'Fully considered', 'Considered very fully'],
    ['Not helpful at all', 'Somewhat helpful', 'Helpful', 'Very helpful'],
    ['Not at all severe', 'Somewhat severe', 'Severe', 'Highly severe']
]

# Calculate max count for y-axis scaling
max_count = max(
    df['fairness'].value_counts().max(),
    df['government_consideration'].value_counts().max(),
    df['policy_helpfulness'].value_counts().max(),
    df['waste_severity'].value_counts().max()
)

for i, var in enumerate(variables):
    plt.subplot(2, 2, i + 1)
    sns.countplot(x=var, data=df, order=orders[i], palette='coolwarm')
    plt.xlabel(f"{titles[i]}\n" + ", ".join([f"{orders[i][j]}={labels[i][j]}" for j in range(len(orders[i]))]))
    plt.ylabel('Number of Respondents')
    plt.title(titles[i])
    plt.ylim(0, max_count + 5)

plt.tight_layout()
```

**More Tasks**: 
- To see the living district distribution of respondents (now the living district information is coded by 0/1, you may consult with Github Copilot the steps to do this task)
- To generate a cross table showing the percentage of respondents in each district who report seeing a food waste bin (`food_waste_behavior`)

```python
# Write code to generate the living district distribution of respondents (now the living district information is coded by 0/1)
# Step 1. Count the number of columns which contain the string "HongKongDistrict" in df

hong_kong_district_columns = [col for col in df.columns if 'HongKongDistrict' in col]
num_hong_kong_district_columns = len(hong_kong_district_columns)
print(f'Number of columns containing "HongKongDistrict": {num_hong_kong_district_columns}')

# Step 2. Combine all district columns into one Series and count frequencies
district_freq = []
for col in hong_kong_district_columns:
    # Extract district name after the underscore
    district_name = col.split('_')[-1]
    # Sum the values (number of participants for this district)
    freq = df[col].sum()
    district_freq.append({'district': district_name, 'frequency': freq})

district_freq_df = pd.DataFrame(district_freq)
print(district_freq_df)

# Step 3. Generate a bar chart to visualize the frequency table above, ordered by frequency descending
plt.figure(figsize=(12, 6))
district_freq_df_sorted = district_freq_df.sort_values(by='frequency', ascending=False)
sns.barplot(x='district', y='frequency', data=district_freq_df_sorted, palette='viridis')
plt.title('Frequency of Participants by Hong Kong District')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
# To generate a cross table showing the percentage of respondents in each district who report seeing a food waste bin (`food_waste_behavior`) The district information is coded by 0/1, for example, HongKongDistrict_CW	HongKongDistrict_KowloonCity	HongKongDistrict_Other	HongKongDistrict_North	HongKongDistrict_Southern. The value of food waste behavior are never_seen, seen_not_used, seen_and_used.
# Step 1. Create a DataFrame to hold district and food waste behavior
district_food_waste = []   
for col in hong_kong_district_columns:
    district_name = col.split('_')[-1]
    # Filter respondents living in this district
    district_respondents = df[df[col] == 1]
    total_respondents = len(district_respondents)
    if total_respondents > 0:
        # Count food waste behavior
        food_waste_counts = district_respondents['food_waste_behavior'].value_counts(normalize=True) * 100
        for behavior in ['never_seen', 'seen_not_used', 'seen_and_used']:
            percentage = food_waste_counts.get(behavior, 0)
            district_food_waste.append({
                'district': district_name,
                'food_waste_behavior': behavior,
                'percentage': percentage
            })
district_food_waste_df = pd.DataFrame(district_food_waste)   
print(district_food_waste_df)

# Step 2. Create a pivot table for better visualization
pivot_table = district_food_waste_df.pivot(index='district', columns='food_waste_behavior', values='percentage').fillna(0)
print(pivot_table)

# Step 3. Visualize the pivot table using a grouped bar chart
pivot_table.plot(kind='bar', figsize=(12, 6), colormap='viridis')
plt.title('Percentage of Respondents by District and Food Waste Behavior')          
plt.ylabel('Percentage (%)')
plt.xlabel('District')
plt.xticks(rotation=45)
plt.legend(title='Food Waste Behavior', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()          




```
